```python
import pandas as pd
import numpy as np

rs = 42
```

```python
df = pd.read_csv('small.csv')
df = df.sample(frac=1, random_state = rs)
df
```

```python
for i in df.columns:
    print(i)
```

```python
cols = ['AGE', 
        'YEARS_WITH_COMPANY',
        'DAILY_KILOMETERS',
        'YEARS_LICENSED',
        'VEHICLE_YEAR',
        'NUMBER_OF_DRIVERS',
        'NUMBER_OF_VEHICLES',
        'DEDUCTIBLE_DCPD',
        'AT_FAULT_CLAIMS',
        'NOT_AT_FAULT_CLAIMS',
        'MINOR_CONVICTIONS',
        'MAJOR_CONVICTIONS',
        'SERIOUS_CONVICTIONS',
        'INCURRED_LOSS_DCPD',
        'GENDER',
        'DRIVER_MARTIAL_STATUS',
        'DRIVER_TRAINING_IND',
        'VEHICLE_AGE']
df = df[cols]
df
```

```python
for i in df:
    print(i)
    print(df[i].isna().sum())
```

```python
for i in df:
    print(i)
    print(df[i].isna().sum()/len(df))
```

```python
for i in df:
    df[i].fillna(method = 'ffill', inplace=True)
    df[i].fillna(method = 'bfill', inplace=True)
```

```python
for i in df:
    print(i)
    print(df[i].isna().sum()/len(df))
```

```python
df
```

```python
for i in df:
    print(i)
    print(df[i].value_counts())
```

```python
df.GENDER = df.GENDER.apply(lambda x: 1 if x == 'M' else 0)
df.DRIVER_MARTIAL_STATUS = df.DRIVER_MARTIAL_STATUS.apply(lambda x: 1 if x == 'S' else 0)
df.DRIVER_TRAINING_IND = df.DRIVER_TRAINING_IND.apply(lambda x: 1 if x == 'Y' else 0)
df.VEHICLE_AGE = df.DRIVER_TRAINING_IND.apply(lambda x: max(x,0))

for i in df:
    print(i)
    print(df[i].value_counts())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

y = df['INCURRED_LOSS_DCPD']
X = pd.DataFrame(scaler.fit_transform(df), columns=df.columns).drop(labels='INCURRED_LOSS_DCPD',axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state = rs)

X_val, X_test, y_val, y_test = train_test_split(X_test,y_test, test_size=0.5, random_state = rs)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV

params = {
        'max_depth': [2, 3, 5, 7, 9, 11],
        'n_estimators': [1000],
        'min_child_weight': [100, 500, 1000, 5000, 10000], 
        'colsample_bytree': [0.4, 0.6, 0.8], 
        'subsample': [0.4, 0.6, 0.8], 
        'learning_rate': [0.1, 0.2, 0.3, 0.4],    
        'random_state': [rs]
    }

xgb = GridSearchCV(XGBRegressor(), cv=3, param_grid=params, verbose=2, n_jobs=-1)


xgb.fit(
    X_train, 
    y_train, 
    eval_metric="rmse", 
    eval_set=[(X_train, y_train), (X_val, y_val)], 
    verbose=True, 
    early_stopping_rounds = 10)

```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

y_pred = xgb.best_estimator_.predict(X_test)

rmse = sqrt(mean_squared_error(y_pred,y_test))

rmse
```

```python
from pickle import dump

with open('xgb.pkl','wb') as f:
    dump(xgb.best_estimator_,f)
```
